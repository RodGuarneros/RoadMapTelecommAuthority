```python
import pandas as pd
import numpy as np
import plotly as plt
import plotly.graph_objects as go
import os
import glob as gb
```

```python
# Reading files function

# use glob to get all the csv files 
# in the folder

path = r'D:/Users/rodrigo.guarneros/Documents/2021/Hoja de Ruta/NubeDeProyectos/Comentarios de los Comisionados/marzo2022/RoadMapTelecommAuthority/Datos_area' 
# use your path
all_files = gb.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, encoding='Latin1')
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

```
